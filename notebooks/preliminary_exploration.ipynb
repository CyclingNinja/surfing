```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

```

```python
pil_im = Image.open('../data/chicama_sample_1.jpg').convert('L')
pil_im
```

```python
# using matplotlib with it
im = np.array(pil_im) 

plt.imshow(im)

x = [100, 200, 300, 400]
y = [200, 500, 200, 400]

plt.plot(x, y, 'r*')
```

```python
# contouring examples
im_contour = np.array(Image.open('../data/chicama_sample_1.jpg').convert('L'), 'f')

fig = plt.figure()
plt.gray()
plt.contour(im_contour, origin='image')
plt.axis('equal')
plt.axis('off')
```

```python
# can pull an histogram inspection of them out of this
fig, (ax1, ax2) = plt.subplots(1,2)
ax1.contour(im_contour, origin='image')
ax1.axis('equal')
ax1.axis('off')

ax2.hist(im_contour.flatten(), 128)
```
